```python
import paramiko
import io
import os
from dotenv import find_dotenv, load_dotenv

load_dotenv(find_dotenv(".env.secrets"), override=True)

SFTP_HOST       = os.getenv("SFTP_HOST")
SFTP_PORT       = int(os.getenv("SFTP_PORT", 22))
SFTP_USER       = os.getenv("SFTP_USER")
SFTP_PASSPHRASE = os.getenv("SFTP_PASSPHRASE")
SFTP_USER   = os.getenv("SFTP_USERNAME")

# ── Reconstruction fiable de la clé ─────────────────────────
# Dans .env.secret, stocker la clé en base64 sur une ligne
import base64
PRIVATE_KEY = base64.b64decode(os.getenv("SFTP_PRIVATE_KEY_B64")).decode("utf-8")

key = paramiko.Ed25519Key.from_private_key(
    io.StringIO(PRIVATE_KEY),
    password=SFTP_PASSPHRASE
)

ssh = paramiko.SSHClient()
ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
ssh.connect(hostname=SFTP_HOST, port=SFTP_PORT, username=SFTP_USER, pkey=key)
sftp = ssh.open_sftp()

print("✅ Connexion SFTP établie")
```

```python
# Cellule notebook — afficher votre clé publique à transmettre à l'admin
key = paramiko.Ed25519Key.from_private_key(io.StringIO(PRIVATE_KEY), password=SFTP_PASSPHRASE)

public_key = f"ssh-ed25519 {key.get_base64()}"
print("Clé publique à déposer sur le serveur :")
print(public_key)
```

```python
# Si cette cellule s'exécute sans erreur → passphrase OK
key = paramiko.Ed25519Key.from_private_key(io.StringIO(PRIVATE_KEY), password=SFTP_PASSPHRASE)
print("✅ Clé déchiffrée avec succès")
print("Fingerprint :", key.get_fingerprint().hex())
```
